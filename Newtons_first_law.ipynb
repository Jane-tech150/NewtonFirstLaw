Newtons law of inertia states that an object at rest will stay at rest, and an object in motion will continue to stay in motion, and a force is required to start move an object, stop it, or change its direction and speed.

```python
import numpy as np
import matplotlib.pyplot as plt

x=0
v=int(input("Enter velocity of the object: "))
t=10
dt=1

position = []
time = []
velocities = []

for i in range (0,t+dt,dt):
  position.append(x)
  time.append(i)
  velocities.append(v)

  x = x+(v*dt)
print(position)
print(time)
print(velocities)

plt.plot(time, position)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Newton's First Law: Postion")
plt.show()

plt.plot(time, velocities)
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.title("Newton's First law: Constant Velocity")
plt.show()
```
